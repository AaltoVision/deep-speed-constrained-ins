```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from matplotlib2tikz import save as tikz_save
```

```python
plt.clf()
fig = plt.figure(figsize=(6,6))
pred=[]
sp=[]
dat_lab=[]
R=[]
with open('spread.csv', 'rb') as csvfile:
     spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
     for row in spamreader:
        pred.append(float(row[0]))
        sp.append(float(row[1]))
        dat_lab.append(int(row[2]))
        R.append([float(row[3]),float(row[4])])
        #print(type(R[-1]))
#plt.plot(sp,pred,'b.',c=[0.5,0.5,0.5])
#plt.xlabel('Ground Truth (m/s)')
#plt.ylabel('Prediction (m/s)')
#axes=plt.gca()
#plt.plot([0,1.6],[0,1.6],c=[0,0,0])
#axes.set_xlim((0.0,1.6))
#axes.set_ylim([0.0,1.6])
#plt.show()
#tikz_save('spread.tex') 

stat=[]
stair=[]
walk=[]
esc=[]
ele=[]

Rstat=[]
Rstair=[]
Rwalk=[]
Resc=[]
Rele=[]

for i in range(0,len(dat_lab)):
    if dat_lab[i]==0:
        stat.append([sp[i],pred[i]])
        Rstat.append(np.array((R[i])))
    elif dat_lab[i]==1:
        walk.append([sp[i],pred[i]])
        Rwalk.append(np.array(R[i]))
    elif dat_lab[i]==2:
        stair.append([sp[i],pred[i]])
        Rstair.append(np.array(R[i]))
    elif dat_lab[i]==3:
        esc.append([sp[i],pred[i]])
        Resc.append(np.array(R[i]))
    else:
        ele.append([sp[i],pred[i]])
        Rele.append(np.array(R[i]))

msize=10
plt.figure()

plt.plot([0,1.5],[0,1.5],'--k')

test=np.array(walk)
#plt.plot(test[:,0],test[:,1],'.',label='Walk',markersize=msize,markerfacecolor='b',
#             markeredgewidth=0.5, markeredgecolor='w')
plt.scatter(test[:,0],test[:,1], s=msize/2, c='g', alpha=0.5, marker='o',edgecolors='w',label='Walking')

test=np.array(stat)
#plt.plot(test[:,0],test[:,1],'.',label='Static',markersize=msize,markerfacecolor='g',
#             markeredgewidth=0.5, markeredgecolor='w')
plt.scatter(test[:,0],test[:,1], s=msize/2, c='b', alpha=0.5, marker='o',edgecolors='w',label='Static')

test=np.array(stair)
#plt.plot(test[:,0],test[:,1],'*',label='Stair',markersize=1.2*msize,markerfacecolor='r',
#             markeredgewidth=0.5, markeredgecolor='w')
plt.scatter(test[:,0],test[:,1], s=msize, c='r', alpha=0.5, marker='s',edgecolors='w',label='Stairs')

test=np.array(esc)
#plt.plot(test[:,0],test[:,1],'+',label='Escalator',markersize=1.2*msize,markerfacecolor='k',
#             markeredgewidth=1.0, markeredgecolor='k')
plt.scatter(test[:,0],test[:,1], s=msize, c='k', alpha=0.5, marker='*',edgecolors='w',label='Escalator')

test=np.array(ele)
#plt.plot(test[:,0],test[:,1],'v',label='Elevator',markersize=1.2*msize,markerfacecolor='k',
#             markeredgewidth=1.0, markeredgecolor='k')
plt.scatter(test[:,0],test[:,1], s=msize, c='c', alpha=0.5, marker='v',edgecolors='w',label='Elevator')


plt.plot([0,1.5],[0,1.5],'--k')
plt.xlabel('Ground truth (m/s)')
plt.ylabel('Prediction (m/s)')

axes=plt.gca()

axes.set_xlim((0.0,1.5))
axes.set_ylim([0.0,1.5])
axes.legend()
#axes.grid(b=True, which='major', color='k', linestyle='--')
tikz_save('scatter.tex',
    figureheight='\\figureheight',
    figurewidth='\\figurewidth') 
plt.show()
```

```python
plt.clf()
pred=[]
sp=[]
t=[]
with open('test.csv', 'rb') as csvfile:
     spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
     for row in spamreader:
        t.append(float(row[0]))
        pred.append(float(row[1]))
        sp.append(float(row[2]))
mint=(np.min(np.array(t)))
plt.plot(t-mint,sp,label='Ground Truth (m/s)',c=[0.7,0.7,0.7])
plt.plot(t-mint,pred,label='Prediction (m/s)',c=[0.2,0.2,0.2])
axes=plt.gca()
plt.ylabel('Speed (m/s)')
plt.xlabel('Time (s)')
axes.set_ylim([0.0,2])
plt.legend()
plt.show()
#tikz_save('test.tex') 
plt.clf()
print(np.sqrt(np.mean(np.square(np.array(pred)-np.array(sp)))))
test_lab=[]
test_lab.append([0,0.28])
test_lab.append([1,0.45])
test_lab.append([2,0.59])
test_lab.append([1,1.55])
test_lab.append([2,2.10])
test_lab.append([1,3.34])
test_lab.append([2,3.48])
test_lab.append([0,4.04])
test_lab.append([2,4.22])
test_lab.append([1,5.19])
test_lab.append([0,5.22])
test_lab.append([1,5.51])
test_lab.append([0,-1])


dat=[]   
dat.append([-1,0])
for label in test_lab:
    tim=label[1]
    dat.append([np.round(np.floor(tim)*60+(tim-np.floor(tim))*100),label[0]])
dat[-1]=[10000,0]
label=[]
start=mint
for i in range(0,len(t)):
    sec=t[i]-start
    for j in range(0,len(dat)-1):
        if sec<dat[j+1][0] and sec>dat[j][0]:
            label.append(dat[j+1][1])
#print(len(data))
print(len(label))
#plt.plot(label)

current=0
green=[0,0.9,0]
red=[0.9,0,0]
blue=[0,0,0.9]
black=[0.2,0.2,0.2]
cyan=[0,0.2,0.2]

for point in dat[0:-1]:
    if point[1]==0:
        plt.axvspan(current, point[0], facecolor=blue,alpha=0.2)
    elif point[1]==1:
        plt.axvspan(current, point[0], facecolor=green,alpha=0.2)
    elif point[1]==2:
        plt.axvspan(current, point[0], facecolor=red,alpha=0.2)
    else:
        plt.axvspan(current, point[0], facecolor=cyan,alpha=0.2)
        
            
        
    current=point[0]

plt.axvspan(current, t[-1]-mint, facecolor=blue,alpha=0.1)
plt.plot(t-mint,sp,label='Ground Truth (m/s)',c=[0.5,0.5,0.5])
plt.plot(t-mint,pred,label='Prediction (m/s)',c=[0.1,0.1,0.1])
axes=plt.gca()
plt.ylabel('Speed (m/s)')
plt.xlabel('Time (s)')
axes.set_ylim([0.0,2])
plt.legend()
#plt.show()
tikz_save('test.tex',
    figureheight='\\figureheight',
    figurewidth='\\figurewidth')
plt.show()

    
```
